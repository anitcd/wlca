## Prevalence of Conditions in Data

To calculate the prevalence of long-term conditions from a dataset, use the following Python code. Update the condition list and the path to the data as needed.

```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

outFile="./data/Prevalence_patientsRandom1K.png"

# Define the list of conditions in the desired order
conditions_to_plot = ['Hypertension', 'CHD', 'StrokeTIA', 'CKD', 'PeripheralVascularDisease', 'AtrialFibrillation', 'HeartFailure', 'Diabetes', 'COPD', 'Asthma', 'Bronchiectasis', 'PainfulCondition', 'Depression', 'Anxiety', 'SchizophreniaBipolar', 'Dementia', 'EatingDisorders', 'LearningDisability', 'AlcoholProblems', 'SubstanceProblems', 'ThyroidDisorders', 'InflammatoryArthritis', 'HearingImpairment', 'VisualImpairment', 'RecentCancer', 'Dyspepsia', 'IrritableBowelSyndrome', 'Constipation', 'DiverticularDisease', 'InflammatoryBowelDisease', 'ChronicSinusitis', 'ViralHepatitis', 'ChronicLiverDisease', 'ProstateDisorders', 'Glaucoma', 'Epilepsy', 'Migraine', 'ParkinsonsDisease', 'MultipleSclerosis', 'PsoriasisEczema']

# Initialize an empty DataFrame to store the data
combined_data = pd.DataFrame()

# Iterate over the CSV files
file_paths = glob.glob('./data/p/*.csv')  # Update with your file paths
for file_path in file_paths:
    data = pd.read_csv(file_path, usecols=['PatientID', 'Age'] + conditions_to_plot)
    combined_data = combined_data.append(data, ignore_index=True)

# Calculate the prevalence of each condition across all files
prevalence = combined_data[conditions_to_plot].mean() * 100  # Calculate the mean and convert to percentage

# Create a bar plot
plt.figure(figsize=(14, 6))
prevalence.plot(kind='bar', color='grey')
plt.title('Prevalence of Conditions', fontsize=14)
plt.xlabel('Conditions', fontsize=14)
plt.ylabel('Prevalence (%)', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=16)  # Rotate x-axis labels for readability
plt.yticks(fontsize=16)
plt.tight_layout()

# Show the plot or save it to a file
plt.savefig(outFile, format='png', bbox_inches='tight', dpi=100)
plt.show()
```

```python
# Rearrange (such as alphabetically or any desired order) the columns of a cleaned LCA/WLCA theta matrix (without '[', ']')
# Update the lists 'original_order' and 'desired_order' as required

import pandas as pd

inFile="./output/WLCAtheta_patients_random_1K_nCluster=5_r1_cleaned.txt"
outFile="./output/WLCAtheta_patients_random_1K_nCluster=5_r1_cleaned_colRearranged.txt"

# Load the matrix from the .txt file
# df = pd.read_csv(inFile, header=None, sep=' ')  # Assuming tab-separated values, change separator if needed
df = pd.read_csv(inFile, header=None, sep='\s+')

# Define the original assumed order of conditions
original_order = [
    'Hypertension', 'CHD', 'StrokeTIA', 'CKD', 'PeripheralVascularDisease', 'AtrialFibrillation',
    'HeartFailure', 'Diabetes', 'COPD', 'Asthma', 'Bronchiectasis', 'PainfulCondition', 'Depression',
    'Anxiety', 'SchizophreniaBipolar', 'Dementia', 'EatingDisorders', 'LearningDisability',
    'AlcoholProblems', 'SubstanceProblems', 'ThyroidDisorders', 'InflammatoryArthritis', 'HearingImpairment',
    'VisualImpairment', 'RecentCancer', 'Dyspepsia', 'IrritableBowelSyndrome', 'Constipation',
    'DiverticularDisease', 'InflammatoryBowelDisease', 'ChronicSinusitis', 'ViralHepatitis', 'ChronicLiverDisease',
    'ProstateDisorders', 'Glaucoma', 'Epilepsy', 'Migraine', 'ParkinsonsDisease', 'MultipleSclerosis', 'PsoriasisEczema'
]

# Define the desired order of conditions (e.g., alphabetical)
desired_order = [
    'AlcoholProblems', 'Anxiety', 'Asthma', 'AtrialFibrillation', 'Bronchiectasis', 'CHD', 'CKD', 'COPD',
    'ChronicLiverDisease', 'ChronicSinusitis', 'Constipation', 'Dementia', 'Depression', 'Diabetes',
    'DiverticularDisease', 'Dyspepsia', 'EatingDisorders', 'Epilepsy', 'Glaucoma', 'HearingImpairment',
    'HeartFailure', 'Hypertension', 'InflammatoryArthritis', 'InflammatoryBowelDisease', 'IrritableBowelSyndrome',
    'LearningDisability', 'Migraine', 'MultipleSclerosis', 'PainfulCondition', 'ParkinsonsDisease',
    'PeripheralVascularDisease', 'ProstateDisorders', 'PsoriasisEczema', 'RecentCancer', 'SchizophreniaBipolar',
    'StrokeTIA', 'SubstanceProblems', 'ThyroidDisorders', 'ViralHepatitis', 'VisualImpairment'
]
    
# Rearrange the columns based on the desired order
df_rearranged = df[[original_order.index(condition) for condition in desired_order]]

# Save the updated matrix back to the .txt file (optional)
df_rearranged.to_csv(outFile, header=False, index=False, sep=' ')

# Now df_rearranged contains the matrix with columns sorted alphabetically as per your desired order
# print(df_rearranged)
print(outFile + ' created.')
```
